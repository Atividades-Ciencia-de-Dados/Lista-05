## C) Agrupe as linhas com postagens iguais ou extremamente semelhantes. Aqui você pode utilizar uma métrica de semelhança de textos. Crie uma variável para representar a quantidade de vezes que a mensagem foi compartilhada. Observe que ao agrupar linhas que possuem a “mesma” postagem (texto), você deve escolher como valor para as variáveis data e hora da postagem, os valores da cópia mais antiga.

```python
import pandas as pd
import numpy as np
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
from sklearn.metrics.pairwise import cosine_similarity

# Vetorização TF-IDF com stopwords em português
stopwords_pt = stopwords.words('portuguese')
vectorizer = TfidfVectorizer(stop_words=stopwords_pt)

# Preencher textos vazios
df['text_content_anonymous'] = df['text_content_anonymous'].fillna('')

# Aplicar TF-IDF
X_tfidf = vectorizer.fit_transform(df['text_content_anonymous'])

# Agrupamento com DBSCAN baseado em similaridade de texto
# DBSCAN exige uma distância: usamos (1 - similaridade do cosseno)
db = DBSCAN(eps=0.3, min_samples=2, metric='cosine')
clusters = db.fit_predict(X_tfidf)

df['grupo_mensagem'] = clusters

# Criar coluna de quantidade de compartilhamentos (mensagens semelhantes no mesmo grupo)
df['qtd_compartilhamentos'] = df.groupby('grupo_mensagem')['grupo_mensagem'].transform('count')

# Converter data da mensagem para datetime
df['date_message'] = pd.to_datetime(df['date_message'], errors='coerce')
```

Classe para transformar textos usando Word2Vec

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import mean_squared_error
from sklearn.base import BaseEstimator, TransformerMixin
from scipy.sparse import hstack, issparse
from gensim.models import Word2Vec

# Classe para transformar textos usando Word2Vec
class Word2VecTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, size=100, window=5, min_count=1, workers=4):
        self.size = size
        self.window = window
        self.min_count = min_count
        self.workers = workers

    def fit(self, X, y=None):
        self.w2v_model = Word2Vec(
            sentences=[text.split() for text in X],
            vector_size=self.size,
            window=self.window,
            min_count=self.min_count,
            workers=self.workers
        )
        return self

    def transform(self, X):
        return np.array([
            np.mean(
                [self.w2v_model.wv[word] for word in text.split() if word in self.w2v_model.wv] or
                [np.zeros(self.size)],
                axis=0
            )
            for text in X
        ])
```

Defini os vetorizadores e modelos, com unigramas, bigramas e trigramas.

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Definir os vetorizadores e modelos
vectorizers = {
    'BOW Unigram': CountVectorizer(ngram_range=(1, 1)),
    'BOW Bigram': CountVectorizer(ngram_range=(1, 2)),
    'BOW Trigram': CountVectorizer(ngram_range=(1, 3)),
    'TF-IDF Unigram': TfidfVectorizer(ngram_range=(1, 1)),
    'TF-IDF Bigram': TfidfVectorizer(ngram_range=(1, 2)),
    'TF-IDF Trigram': TfidfVectorizer(ngram_range=(1, 3)),
    'Word2Vec': Word2VecTransformer()
}
```

Função para executa: Treinamento, Validação e Teste com Outer K-Fold Cross-Validation. A função permite você dar com entrada colunas de texto e numéricas.

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error
from scipy.sparse import issparse, hstack
import numpy as np

def evaluate_model_with_outer_cv_text_only_vectorizer(X_train_val, y_train_val, X_test, y_test,
                                                       vectorizer, model, scaler=None, normalizer=None,
                                                       texto_coluna='text_content_anonymous',
                                                       colunas_numericas=None,
                                                       colunas_continuas=None,
                                                       outer_splits=5):
    outer_cv = KFold(n_splits=outer_splits, shuffle=True, random_state=42)
    results = []

    for train_index, val_index in outer_cv.split(X_train_val):
        X_train, X_val = X_train_val.iloc[train_index], X_train_val.iloc[val_index]
        y_train, y_val = y_train_val.iloc[train_index], y_train_val.iloc[val_index]

        # Vetorizar o texto
        X_train_text_vec = vectorizer.fit_transform(X_train[texto_coluna])
        X_val_text_vec = vectorizer.transform(X_val[texto_coluna])
        X_test_text_vec = vectorizer.transform(X_test[texto_coluna])

        # Verifica se o vetor é esparso ou denso
        is_sparse = issparse(X_train_text_vec)

        # Normaliza vetores de texto se for denso e houver normalizador
        if normalizer and not is_sparse:
            X_train_text_vec = normalizer.fit_transform(X_train_text_vec)
            X_val_text_vec = normalizer.transform(X_val_text_vec)
            X_test_text_vec = normalizer.transform(X_test_text_vec)

        # As colunas numericas são todas inclusive as booleanas que converti pra int (0 ou 1)
        if colunas_numericas:
            X_train_num = X_train[colunas_numericas].copy()
            X_val_num = X_val[colunas_numericas].copy()
            X_test_num = X_test[colunas_numericas].copy()

            # Aplica Normalização apenas se for fornecido o scaler e as colunas contínuas
            if scaler and colunas_continuas:
                scaler.fit(X_train_num[colunas_continuas])
                X_train_num[colunas_continuas] = scaler.transform(X_train_num[colunas_continuas])
                X_val_num[colunas_continuas] = scaler.transform(X_val_num[colunas_continuas])
                X_test_num[colunas_continuas] = scaler.transform(X_test_num[colunas_continuas])

            if is_sparse:
                X_train_final = hstack([X_train_text_vec, X_train_num.values])
                X_val_final = hstack([X_val_text_vec, X_val_num.values])
                X_test_final = hstack([X_test_text_vec, X_test_num.values])
            else:
                X_train_final = np.hstack([X_train_text_vec, X_train_num.values])
                X_val_final = np.hstack([X_val_text_vec, X_val_num.values])
                X_test_final = np.hstack([X_test_text_vec, X_test_num.values])
        else:
            # Se for fornecido só a coluna de texto executa apenas essa parte
            if is_sparse:
                X_train_final = X_train_text_vec
                X_val_final = X_val_text_vec
                X_test_final = X_test_text_vec
            else:
                X_train_final = np.array(X_train_text_vec)
                X_val_final = np.array(X_val_text_vec)
                X_test_final = np.array(X_test_text_vec)

        # Treina e avalia o modelo
        model.fit(X_train_final, y_train)
        y_val_pred = model.predict(X_val_final)
        y_test_pred = model.predict(X_test_final)

        val_mse = mean_squared_error(y_val, y_val_pred)
        test_mse = mean_squared_error(y_test, y_test_pred)
        val_rmse = np.sqrt(val_mse)
        test_rmse = np.sqrt(test_mse)
        val_mae = mean_absolute_error(y_val, y_val_pred)
        test_mae = mean_absolute_error(y_test, y_test_pred)

        results.append((val_mse, val_rmse, val_mae, test_mse, test_rmse, test_mae))

    return (
        np.mean([r[0] for r in results]),  # val_mse
        np.mean([r[1] for r in results]),  # val_rmse
        np.mean([r[2] for r in results]),  # val_mae
        np.mean([r[3] for r in results]),  # test_mse
        np.mean([r[4] for r in results]),  # test_rmse
        np.mean([r[5] for r in results])   # test_mae
    )
```

- Função para fazer o pré processamento do texto

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
from unidecode import unidecode

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('portuguese'))

def preprocess_text(text):
    def substituir_dominios(texto):
        def extrair_dominio(url):
            dominio = re.sub(r'^https?://(?:www\.)?|www\.', '', url)
            dominio = re.split(r'[/?#]', dominio)[0]
            return dominio.split('.')[0]
        return re.sub(r'https?://(?:www\.)?\S+|www\.\S+', lambda m: extrair_dominio(m.group(0)), texto)

    # Substituir domínios
    text = substituir_dominios(text)

    # Minúsculas e remoção de acentos
    text = unidecode(text.lower())

    # Remover pontuações
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Substituir emojis e risadas repetidas
    text = re.sub(r'([\U00010000-\U0010FFFF])\1+', r'\1', text)
    text = re.sub(r'(ha){2,}', 'haha', text, flags=re.IGNORECASE)
    text = re.sub(r'(ka){2,}', 'kaka', text, flags=re.IGNORECASE)
    text = re.sub(r'k{2,}', 'kk', text, flags=re.IGNORECASE)

    # Reduzir espaços
    text = re.sub(r'\s+', ' ', text).strip()

    # Remover stopwords
    palavras_filtradas = [word for word in text.split() if word not in stop_words]
    text = ' '.join(palavras_filtradas)

    return text
```
